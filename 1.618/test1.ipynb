```python
data_dis = web.DataReader(['DIS'], 'yahoo', start, end)
data_dis.columns= data_dis.columns.swaplevel(0, 1)
data_dis.rename(columns= {'Close': 'close'}, inplace=True)
data_dis.sort_index(axis= 1, level=1, inplace=True)
```

```python
# DIS: Configuring a Monte Carlo simulation to forecast 5 years cumulative returns
test1 = MCSimulation(
    portfolio_data = data_dis,
    num_simulation = 1000,
    num_trading_days = 252*5
)

# Print the simulation input data
test1.portfolio_data.head()
```

```python
#Disney'DIS': Running a Monte Carlo simulation to forecast 5 years cumulative returns
test1.calc_cumulative_return()
```

```python
# DIS: Fetch summary statistics from the Monte Carlo simulation results
dis_MC_summary = test1.summarize_cumulative_return()

# Print summary statistics
print(dis_MC_summary)
```

```python
# TSLA: Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $10,000 investments in TSLA stocks
dis_ci_lower = round(dis_MC_summary[8]*10000,2)
dis_ci_upper = round(dis_MC_summary[9]*10000,2)

# Print results
print(f"There is a 95% chance that an initial investment of $10,000 in the portfolio"
      f" over the next year will end within in the range of"
      f" ${dis_ci_lower} and ${dis_ci_upper}.")
```
